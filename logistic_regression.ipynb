```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/home/hauuu/Desktop/deep learning/DL_Tutorial/L2/dataset.csv").values
data
```

```python
# data.values
N,d=data.shape
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
data
```

```python
x = data[:, 0:d-1].reshape(-1, d-1)
y = data[:, 2].reshape(-1, 1)
```

```python
x_cho_vay=x[y[:,0]==1]
x_tu_choi=x[y[:,0]==0]
```

```python
plt.scatter(x_cho_vay[:, 0], x_cho_vay[:, 1], c='red', edgecolors='none', s=30, label='cho vay')
plt.scatter(x_tu_choi[:, 0], x_tu_choi[:, 1], c='blue', edgecolors='none', s=30, label='từ chối')
plt.legend(loc=1)
plt.xlabel('mức lương (triệu)')
plt.ylabel('kinh nghiệm (năm)')
```

```python
x
```

```python
x=np.hstack((np.ones((N,1)),x))
w=np.array([0.,0.1,0.1]).reshape(-1,1)
```

```python
w
```

```python
numOfItration=1000
cost=np.zeros((numOfItration,1))
learning_rate=0.00001
```

```python
for i in range(1,numOfItration):
    y_predict=sigmoid(np.dot(x,w))
    cost[i]=-np.sum(np.multiply(y,np.log(y_predict))+np.multiply(1-y,np.log(1-y_predict)))
    #Gradient descent
    w=w-learning_rate*np.dot(x.T,y_predict-y)
    print(cost[i])
```

```python
np.save('weight logistic',w)

```

```python
x1=100
x2=1.0
t=0.5
y=w[0]+w[1]*x1+w[2]*x2
if y>=np.log(1/t-1): print("YES")
else: print("NO")
```
